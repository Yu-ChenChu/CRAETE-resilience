### Webscrabing Sunshine Sunset Data
Data Source: https://sunrise-sunset.org/us/los-angeles-ca/2022/1

The data will be used for estimate solar power production.

#### Scrabe the Data for 2022
Produce the Date, Sunrise, and Sunset data.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Create an empty DataFrame to store all the data
combined_df = pd.DataFrame(columns=["Date", "Sunrise", "Sunset"])

# Iterate over the URLs
for i in range(1, 13):
    url = f"https://sunrise-sunset.org/us/los-angeles-ca/2022/{i}"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    table = soup.find('table', {'id': 'month'})
    rows = []
    for tr in table.select('tr')[2:]:
        tds = tr.select('td')
        date = tr.select('span.number')[0].text.strip()
        sunrise = tds[1].text.strip()
        sunset = tds[2].text.strip()   
        rows.append([date, sunrise, sunset])
    # Append the data to the combined DataFrame
    df = pd.DataFrame(rows, columns=["Date", "Sunrise", "Sunset"])
    combined_df = combined_df.append(df, ignore_index=True)

# Save the combined data to a CSV file
combined_df.to_csv('..\\Data\\LA_Sunrise_Sunset_Data\\sunrise_sunset_data.csv', index=False)
```

#### Scrabe the Data from 2018 to 2023
Produce the Date, Sunrise, Sunset, and Day Length data.

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Create an empty DataFrame to store all the data
combined_df = pd.DataFrame(columns=["Date", "Sunrise", "Sunset", "Day_Length" ])

# Iterate over the URLs
for y in range(2018, 2023):
    for i in range(1, 13):
        url = f"https://sunrise-sunset.org/us/los-angeles-ca/{y}/{i}"
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')
        table = soup.find('table', {'id': 'month'})
        rows = []
        for tr in table.select('tr')[2:]:
            tds = tr.select('td')
            date = tr.select('span.number')[0].text.strip()
            sunrise = tds[1].text.strip()
            sunset = tds[2].text.strip()  
            day_length = tds[4].text.strip()  
            rows.append([date, sunrise, sunset, day_length])
        # Append the data to the combined DataFrame
        df = pd.DataFrame(rows, columns=["Date", "Sunrise", "Sunset", "Day_Length"])
        combined_df = combined_df.append(df, ignore_index=True)

    # Save the combined data to a CSV file
    combined_df.to_csv('..\\Data\\LA_Sunrise_Sunset_Data\\2018_2023_sunrise_sunset.csv', index=False)   
```
