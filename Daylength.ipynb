### Calculate Monthly and Seasonal Sunrise Sunset Data

The data will be used for estimate solar power production.

```python
import pandas as pd
import numpy as np

# read the Excel file 
df_6yrs = pd.read_csv('..\\Data\\LA_Sunrise_Sunset_Data\\2018_2023_sunrise_sunset.csv')
df_6yrs
```

```python
# Remove 202303 to 202312
df = df_6yrs.drop(df_6yrs.index[1520:])
# Remove 201801 to 201802
df = df.drop(df.index[:59])
# Subset Data
df = df[['Date','Sunrise','Sunset']]
df
```

```python
# Convert the "Sunshine" column to time
df['Sunrise_time'] = pd.to_datetime(df['Sunrise'], infer_datetime_format=True)
df['Sunset_time'] = pd.to_datetime(df['Sunset'], infer_datetime_format=True)
df
```

```python
# Calculate Daylength
df['DayLength_time'] = df['Sunset_time']-df['Sunrise_time']
df
```

```python
# Add day time
df['Date_time'] = df['Date'].str.slice(5)
df
```

```python
# Delete date only keep month 
df['Date_time'] = df['Date_time'].str.slice(stop = 3)
df
```

```python
# Define a dictionary to map column a values to the corresponding season
season_map = {'Mar': 'spring', 'Apr': 'spring', 'May': 'spring',
              'Jun': 'summer', 'Jul': 'summer', 'Aug': 'summer',
              'Sep': 'fall', 'Nov': 'fall', 'Oct': 'fall',
              'Dec': 'winter', 'Jan': 'winter', 'Feb': 'winter'}

# Use the map() method to apply the season_map dictionary to column a and store the result in a new column called "Season"
df['Season'] = df['Date_time'].map(season_map)
df
```

```python
# Calculate mean sunrise, sunset and daylength
df_month = df.groupby(df['Date_time']).mean()
df_month
```

```python
# round the column to the nearest second
df_month['DayLength_time'] = pd.to_timedelta(np.ceil(df_month['DayLength_time'].dt.total_seconds()).astype(int), unit='s')
df_month['Sunrise_time'] = pd.to_datetime(np.round(df_month['Sunrise_time'].astype(np.int64) / 10**9).astype(int), unit='s')
df_month['Sunset_time'] = pd.to_datetime(np.round(df_month['Sunset_time'].astype(np.int64) / 10**9).astype(int), unit='s')

df_month
```

```python
# Reset the order of months
order = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
df_month = df_month.sort_values(by=['Date_time'], key=lambda x: x.map({v: i for i, v in enumerate(order)}))
df_month
```

```python
# Create sunrise and sunset range
df_month['Sunrise_Range'] = ['06:45:01 to 07:00:00',
                          '06:30:01 to 06:45:00',
                          '06:30:01 to 06:45:00',
                          '06:15:01 to 06:30:00',
                          '05:45:01 to 06:00:00',
                          '05:30:01 to 05:45:00',
                          '05:45:01 to 06:00:00',
                          '06:00:01 to 06:15:00',
                          '06:30:01 to 06:45:00', 
                          '06:45:01 to 07:00:00',
                          '06:30:01 to 06:45:00',
                          '06:45:01 to 07:00:00' ]    

df_month['Sunset_Range']= ['17:00:01 to 17:15:00',
                           '17:30:01 to 17:45:00',
                           '18:30:01 to 18:45:00',
                           '19:15:01 to 19:30:00',
                           '19:45:01 to 20:00:00',
                           '20:00:01 to 20:15:00',
                           '20:00:01 to 20:15:00',
                           '19:30:01 to 19:45:00',
                           '19:00:01 to 19:15:00',
                           '18:15:01 to 18:30:00',
                           '16:45:01 to 17:00:00',
                           '16:45:01 to 17:00:00']

df_month

# export the DataFrame to an Excel file
df_month.to_csv('..\\Data\\LA_Sunrise_Sunset_Data\\sunrise_sunset_month.csv', index=True)
```

```python
# Get the season mean
df_season = df.groupby(df['Season']).mean()
df_season
```

```python
# round the column to the nearest second
df_season['DayLength_time'] = pd.to_timedelta(np.ceil(df_season['DayLength_time'].dt.total_seconds()).astype(int), unit='s')
df_season['Sunrise_time'] = pd.to_datetime(np.round(df_season['Sunrise_time'].astype(np.int64) / 10**9).astype(int), unit='s')
df_season['Sunset_time'] = pd.to_datetime(np.round(df_season['Sunset_time'].astype(np.int64) / 10**9).astype(int), unit='s')

df_season
```

```python
# Create sunrise and sunset range
df_season['Sunrise_Range'] = ['06:45:01 to 07:00:00',
                          '06:15:01 to 06:30:00',
                          '05:45:01 to 06:00:00',
                          '06:45:01 to 07:00:00',]    

df_season['Sunset_Range']= ['18:00:01 to 18:15:00',
                           '19:15:01 to 19:30:00',
                           '19:45:01 to 20:00:00',
                           '17:00:01 to 17:15:00']

df_season

# export the DataFrame to an Excel file
df_season.to_csv('..\\Data\\LA_Sunrise_Sunset_Data\\sunrise_sunset_season.csv', index=True)
```
